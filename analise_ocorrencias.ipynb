```python
import pandas as pd
aeronave = pd.read_csv("aeronave.csv", encoding='latin1', sep = ';')
fator_contribuinte = pd.read_csv("fator_contribuinte.csv", encoding='latin1', sep = ';')
ocorrencia = pd.read_csv("ocorrencia.csv", encoding='latin1', sep = ';')
ocorrencia_tipo = pd.read_csv("ocorrencia_tipo.csv", encoding='latin1', sep = ';')
recomendacao = pd.read_csv("recomendacao.csv", encoding='latin1', sep = ';')
```

```python
ocorrencia.head()
```

```python
import sqlite3
con = sqlite3.connect("ocorrencias.db")

ocorrencia.to_sql("tb_ocorrencia", con)
aeronave.to_sql("tb_aeronave", con)
fator_contribuinte.to_sql("tb_fator_contribuinte", con)
ocorrencia_tipo.to_sql("tb_ocorrencia_tipo", con)
recomendacao.to_sql("tb_recomendacao", con)
```

```python
import pandas as pd
import sqlite3

con = sqlite3.connect("ocorrencias.db")

comando_sql = "SELECT ocorrencia_tipo_categoria, COUNT(ocorrencia_tipo_categoria) FROM tb_ocorrencia_tipo GROUP BY ocorrencia_tipo_categoria"

dados = pd.read_sql(comando_sql, con)

display(dados)
```

```python
import sqlite3

con = sqlite3.connect("ocorrencias.db")
cursor = con.cursor()

comando_sql = "DELETE FROM tb_ocorrencia_tipo WHERE ocorrencia_tipo_categoria='***';"

cursor.execute(comando_sql)
con.commit()
```
